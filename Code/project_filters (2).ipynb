Fourier (under experiment)


```python

height, width = imgreal.shape

 
complexer = np.zeros((height, width), dtype=complex)

  
for i in range(height):
  for j in range(width):
    complexer[i, j] = imgreal[i, j] * np.exp(-1j * 2 * np.pi * (i / height) * (j / width))

cv2_imshow(complexer)
```

```python
fourierforward = np.empty((height, width), dtype=complex)
for u in range(height):
  for v in range(width):
    fourierforward[u, v] = 0
    for i in range(height):
      for j in range(width):
        fourierforward[u, v] += complexer[i, j] * np.exp(1j * 2 * np.pi * (i * u / height) + 1j * 2 * np.pi * (j * v / width))

  

cv2_imshow(fourierforward)
```
