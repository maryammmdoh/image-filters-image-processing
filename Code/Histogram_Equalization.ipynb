```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
import math
```

```python
img = cv2.imread("/content/png-transparent-hummingbird-free-buckle-hummingbird-free-logo-design-template-computer-fauna-thumbnail.png",0)
img.shape
size=img.shape[0]*img.shape[1]
size
```

```python
img2 = img.copy()
img3 = img.copy()
```

```python
frequency=[0.0]*256
print(frequency)
for r in range(img.shape[0]):
  for c in range(img.shape[1]):
    frequency[img[r,c]] += 1
print(frequency)
```

```python
probability=frequency.copy()
i=0
while(i<256):
  probability[i]=probability[i]/129600
  i+=1
print(probability)
```

```python
normalized=[0]*256
i=0
sum=0
while(i<256):
  sum+=probability[i]
  normalized[i]+= 255*sum
  i+=1
print(normalized)
i=0
while(i<256):
  normalized[i] = np.round (normalized[i])
  i+=1
print(normalized)
```

```python
for r in range(img.shape[0]):
  for c in range(img.shape[1]):
    img2[r,c]=normalized[img[r,c]]
cv2_imshow(img)
cv2_imshow(img2)
```

```python
plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.show()
plt.hist(img2.flatten(),256,[0,256], color = 'r')
plt.show()
```
